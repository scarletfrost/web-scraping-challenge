```python
# Declare Dependencies 
from bs4 import BeautifulSoup
from splinter import Browser
import pandas as pd
import requests
```

```python
#path to the chromedriver
executable_path = {"executable_path": "chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)
```

NASA Mars News

```python
NASA_Mars_News_url = 'https://mars.nasa.gov/news/'
browser.visit(NASA_Mars_News_url)

html = browser.html
# Parse HTML with Beautiful Soup
results = BeautifulSoup(html, 'html.parser')
print(results.body.prettify())
```

```python
#collectting the latest the News Title and Paragraph Text

title_news = results.find_all('div', class_='content_title')

news_title = title_news[4].text

print('News Title:', news_title )
```

```python
#skipping to get the actual news text seen in the browser
para = results.find_all('div', class_="rollover_description")
para_news = para[9].text
print('Paragraph Text:', para_news)
```

JPL Mars Space Images

```python
JPL_Mars_Space_image_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(JPL_Mars_Space_image_url)

html = browser.html
# Parse HTML with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')

print(soup.body.prettify())
```

```python
f_img_url=soup.find("div", class_="carousel_items")
print(f_img_url.prettify())
```

```python
#getting featured image url

bg_img_url=soup.find('article')['style']
bg_img_url_split=bg_img_url.split("'")
bg_img_url_f=bg_img_url_split[1]
bg_img_url_f
```

```python
#printting featured image full url
base='https://www.jpl.nasa.gov'

featured_image_url = base + bg_img_url_f

print('featured_image_url = ' + featured_image_url)
```

Mars Weather

```python
Mars_Weather_url = 'https://twitter.com/marswxreport?lang=en'
response = requests.get(Mars_Weather_url)

# Create BeautifulSoup object
twitter_soup = BeautifulSoup(response.text, 'lxml')

print(twitter_soup.body.prettify())
```

```python
t_data = twitter_soup.find_all('div', class_='js-tweet-text-container')
t_data
```

```python
t_text = twitter_soup.find_all('p',class_='TweetTextSize TweetTextSize--normal js-tweet-text tweet-text')
t_text
```

```python
# loop through to get tweets that start with 'InSight' to get the weather
w_text = 'InSight '

for t in t_text:
    if w_text in t.text:
        mars_weather = t.text
print('mars_weather = '+mars_weather)
```

Mars Facts

```python
mars_facts_url = "https://space-facts.com/mars/"
facts_data = pd.read_html(mars_facts_url)
facts_data_df = facts_data[0]
facts_data_df
```

```python
#convert the data to a HTML table string and remove \n from HTML string
facts_df_html = facts_data_df.to_html()
facts_df_html = facts_df_html.replace("\n",'')
facts_df_html
```

Mars Hemispheres

```python
hemi_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(hemi_url)
# Create BeautifulSoup object; parse with 'html'
html = browser.html
hem_soup = BeautifulSoup(html, 'html.parser')

print(hem_soup.prettify())
```

```python
# Create list to hold all hemisphere titles and image urls dictionaries
hemisphere_image_urls = []
# Create dictionary to hold all hemisphere titles and image urls
hem_dict = {'title': [], 'img_url': [],}

x = hem_soup.find_all('h3')
print(x)
```

```python
#looping through to get the titles and urls
for i in x:
    t = i.get_text()
    title = t.strip('Enhanced')
    browser.click_link_by_partial_text(t)
    hem_url = browser.find_link_by_partial_href('download')['href']
    hem_dict = {'title': title, 'img_url': hem_url}
    
    # Append dictionaries to list
    hemisphere_image_urls.append(hem_dict)
    browser.back()

hemisphere_image_urls
```

loading all found answers to mongo DB

```python
# Imports
import pandas as pd
import pymongo
from pymongo import MongoClient

# creating a dictionary to hold all the data found above  
mission_to_mars = {"news_title":news_title,"news_paragraph":para_news,"featured_image_url":featured_image_url,"mars_weather":mars_weather,"mars_facts":facts_df_html,"hemisphere_image_url":hemisphere_image_urls}

# Connect to MongoDB
conn =  'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client['Mars']
collection = db['mars_data']

# Insert collection
collection.insert_one(mission_to_mars)
```
